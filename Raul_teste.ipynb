```python

# ==========================================================
# ARQUIVO ROOT
# ==========================================================

ROOT_FILE = "root_files/V1-Hrec_hoft_16384Hz_OMICRON-1243800003-1800004.root"

# ==========================================================
# LEITURA DO ROOT
# ==========================================================

print("Lendo arquivo ROOT...")
tree = uproot.open(ROOT_FILE)["triggers"]
df = tree.arrays(library="pd")

print(f"Número total de triggers: {len(df)}")
print("\nColunas disponíveis:")
print(df.columns)

# ==========================================================
# DURAÇÃO E BANDA
# ==========================================================

df["duration"] = np.maximum(
    df["tend"] - df["tstart"],
    0.05
)

df["bandwidth"] = np.maximum(
    df["fend"] - df["fstart"],
    0.1
)

# ==========================================================
# ENERGIA DO GLITCH
# ==========================================================

df["energy"] = df["snr"]**2

# ==========================================================
# ESTATÍSTICAS GLOBAIS
# ==========================================================

Emin = df["energy"].min()
Emax = df["energy"].max()
Emean = df["energy"].mean()
Emedian = df["energy"].median()

print("\nEstatísticas das energias:")

print(f"Energia mínima : {Emin:.3f}")
print(f"Energia máxima : {Emax:.3f}")
print(f"Energia média  : {Emean:.3f}")
print(f"Energia mediana: {Emedian:.3f}")

# ==========================================================
# MOSTRA PRIMEIROS GLITCHES
# ==========================================================

print("\nPrimeiros glitches:")

print(
    df[
        [
            "time",
            "frequency",
            "snr",
            "energy",
            "duration",
            "bandwidth"
        ]
    ].head(10)
)

# ==========================================================
# HISTOGRAMA DAS ENERGIAS
# ==========================================================

plt.figure(figsize=(8,5))

plt.hist(
    df["energy"],
    bins=100,
    log=True
)

plt.xlabel(r"$E \propto {\rm SNR}^2$")
plt.ylabel("Número de glitches")
plt.title("Distribuição das energias dos glitches")

plt.tight_layout()
plt.show()

# ==========================================================
# SALVA RESULTADOS
# ==========================================================

df.to_csv(
    "glitch_energies.csv",
    index=False
)

print("\nArquivo salvo:")
print("glitch_energies.csv")
```
